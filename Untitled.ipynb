```python
from tkinter import *
from tkinter import ttk
import time


sc = Tk()
sc.title("Thu thap thong tin")
sc.geometry("500x350")

data1 = StringVar()
data2 = StringVar()
data3 = StringVar()
data4 = StringVar()
data5 = StringVar()
data6 = StringVar()
data7 = StringVar()

def nhapdata():
    thongbao = Tk()
    thongbao.title("Thông báo")
    thongbao.geometry("330x100")
    if (data1.get() == "" or 
        data2.get() == ""or 
        data2.get() == ""or 
        data3.get() == ""or 
        data4.get() == ""or 
        data5.get() == ""or
        data6.get() == ""or
        data7.get() == ""):
       
        dongthongbao = Label(thongbao,text="Bạn chưa nhập kìa")
        dongthongbao.pack()
    else:
        chuoidata=(data1.get()+","+data2.get()+","+data3.get()+",\'"+data4.get()+","+
                   data5.get()+","+data6.get()+","+data7.get())
        dongthongbao = Label(thongbao,text="OK Done")
        dongthongbao.pack()
        
        #chuoidata=data1.get()+data2.get()+data3.get()+data4.get()+data5.get()+data6.get()+data7.get()
        f = open("Data.csv","a",encoding="utf8")
        f.write(chuoidata+"\n")
        f.close()
        if data5.get() not in sanpham["values"]:               #giá trị nhập vào không trùng
            f=open("sanpham.txt","a",encoding="utf8")          #mở file sanpham.txt và "a"ppend thêm dữ liệu nhập vào file
            f.write(data5.get()+"\n")                          #viết thêm dữ liệu vào lấy từ ô data5, thêm dấu xuống dòng
            f.close()                                          # luôn phải đóng lệnh f.close()
            f=open("sanpham.txt","r",encoding="utf8")          #mở file sanpham.txt và "r"ead các dữ liệu trong file
            sanpham["values"]=f.readlines()                    #khi đó chuỗi "values" bằng các giá trị đọc trong file
            f.close()                                          # luôn phải đóng lệnh f.close()
            
        if data7.get() not in nganh["values"]:               
            f=open("nganh.txt","a",encoding="utf8")         
            f.write(data7.get()+"\n")                          
            f.close()                                          
            f=open("nganh.txt","r",encoding="utf8")          
            nganh["values"]=f.readlines()                    
            f.close()             
#Xóa trắng các ô nhập giá trị
        data1.set("")
        data2.set("")
        data3.set("")
        data4.set("")
        data5.set("")
        data6.set("")
        data7.set("")
    
#Tên khách hàng,Công ty,Mail,Số điện thoại,Sản phẩm,Giá trị mua,Ngành/Group        
      

Label(sc,text="Nhap thong tin khach hang",fg = "red",font = ("Aria",15)).grid(
    row=1,columnspan = 3,column = 1,padx = 80,pady = (20,40))

Label(sc,text="Tên khách hàng:").grid(row=2,column=1,sticky = W,padx = (30,0))
Label(sc,text="Công ty:").grid(row=3,column=1,sticky = W,padx = (30,0))
Label(sc,text="Mail:").grid(row=4,column=1,sticky = W,padx = (30,0))
Label(sc,text="Số điện thoại:").grid(row=5,column=1,sticky = W,padx = (30,0))
Label(sc,text="Sản phẩm:").grid(row=6,column=1,sticky = W,padx = (30,0))
Label(sc,text="Giá trị mua:").grid(row=7,column=1,sticky = W,padx = (30,0))
Label(sc,text="Ngành/Group:").grid(row=8,column=1,sticky = W,padx = (30,0))

Entry(sc,textvariable = data1,width = 50).grid(row=2,column=2,padx = (0,30))
Entry(sc,textvariable = data2,width = 50).grid(row=3,column=2,padx = (0,30))
Entry(sc,textvariable = data3,width = 50).grid(row=4,column=2,padx = (0,30))
Entry(sc,textvariable = data4,width = 50).grid(row=5,column=2,padx = (0,30))
#Entry(sc,textvariable = data5,width = 50).grid(row=6,column=2,padx = (0,30))
#dùng thư viện ttk, tạo xổ chọn bằng đọc các giá trị trong widget "sanpham" bằng khai báo chuỗi "values"
sanpham=ttk.Combobox(sc,width=47,textvariable = data5)
sanpham.grid(row=6,column=2,padx = (0,30))
#sanpham["values"]=["A","B","C","D"]
#mỗi lần nhập thêm 1 giá trị mới xổ chọn "sanpham", lấy các giá trị trong "values" bằng trong file sanpham.txt đọc được
f=open("sanpham.txt","r",encoding="utf8")
sanpham["values"]=f.readlines()
f.close()

Entry(sc,textvariable = data6,width = 50).grid(row=7,column=2,padx = (0,30))
#Entry(sc,textvariable = data7,width = 50).grid(row=8,column=2,padx = (0,30))
nganh=ttk.Combobox(sc,width=47,textvariable = data7)
nganh.grid(row=8,column=2,padx = (0,30))
#nganh["values"]=["X","Y","Z","W"]
f=open("nganh.txt","r",encoding="utf8")
nganh["values"]=f.readlines()
f.close()

Button(sc,text = "Input", command = nhapdata).grid(column = 2, row = 10)


sc.mainloop()
```
